```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Read the dataset
df = pd.read_csv('/imdb_top_1000.csv')
```

```python
df.head()
```

```python
# Drop unnecessary columns
df.drop(columns=['Poster_Link', 'Overview', 'Series_Title'], inplace=True)
```

```python
# Check for null values
null_counts = df.isnull().sum()
null_columns = null_counts[null_counts != 0].index.tolist()
print({col: count for col, count in null_counts.items() if count != 0})
```

```python
# Feature Engineering
df['Has_Meta_score'] = df['Meta_score'].notnull().astype(int)
df['Certificate'].fillna(df['Certificate'].mode()[0], inplace=True)
df['Meta_score'].fillna(0, inplace=True)
df['Gross'].fillna(df['Gross'].mode()[0], inplace=True)
df['Runtime'] = df['Runtime'].str.split(' ').str[0].astype(int)
df.rename(columns={"Runtime": "Runtime('min')"}, inplace=True)
df['Runtime_Category'] = pd.cut(df["Runtime('min')"], bins=[0, 50, 100, 150, 200, 250, float('inf')],
                                 labels=['Very Short', 'Short', 'Medium', 'Long', 'Very Long', 'Extremely Long'])

df.loc[966, 'Released_Year'] = int((int(df.iloc[965]['Released_Year'])+int(df.iloc[967]['Released_Year']))/2)
df['Released_Year'] = df['Released_Year'].astype(int)
df['Release_Decade'] = (df['Released_Year'] // 10) * 10
df['Gross'] = df['Gross'].str.replace(',', '').astype(int)
df['Budget_Category'] = pd.cut(df['Gross'], bins=[0, 10000000, 50000000, float('inf')], labels=['Low', 'Medium', 'High'])
```

```python
# Data Visualization
plt.figure(figsize=(8, 2))
sns.histplot(df["Runtime('min')"], kde=True, bins=20)
plt.title("Histogram of Runtime")
plt.xlabel("Runtime (minutes)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 2))
plt.hist(df['Gross'], bins=20)
plt.title('Distribution of Gross')
plt.xlabel('Gross')
plt.ylabel('Frequency')
plt.grid(True)
plt.ticklabel_format(style='plain', axis='both')
plt.show()
```

```python
plt.figure(figsize=(7, 3))
sns.countplot(data=df, x='Certificate')
plt.title('Count of Movies by Certificate')
plt.xlabel('Certificate')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=[9,3])
# Calculate mean IMDb rating for each genre
genre_ratings = df.groupby('Genre')['IMDB_Rating'].mean().reset_index()
# Select the top 8 genres based on mean IMDb rating
top_8_genres = genre_ratings.nlargest(8, 'IMDB_Rating')
# Plotting the top 8 genres
bar_plot = sns.barplot(data=top_8_genres, x="Genre", y="IMDB_Rating")
bar_plot.set_xticklabels(bar_plot.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.title('Top 8 Genres by Mean IMDb Rating')
plt.xlabel('Genre')
plt.ylabel('Mean IMDb Rating')
plt.show()
```

```python
# Machine Learning
df = pd.get_dummies(df, drop_first=True)
X = df.drop('IMDB_Rating', axis=1)
Y = df['IMDB_Rating']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.3, random_state=42)
```

```python
# Random Forest Regressor
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(X_train, Y_train)
Y_pred = rf_regressor.predict(X_test)
mse = mean_squared_error(Y_test, Y_pred)
print("Mean Squared Error (Random Forest):", mse)
print('Predicted:', round(rf_regressor.predict(X.iloc[0:1])[0], 1), '\n',
      'Real one:', df.loc[0, 'IMDB_Rating'])
```

```python
# Gradient Boosting Regressor with Hyperparameter Tuning
param_grid_gb = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 5],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'subsample': [0.8, 1.0]
}
gb_regressor = GradientBoostingRegressor(random_state=42)
gb_random = RandomizedSearchCV(estimator=gb_regressor, param_distributions=param_grid_gb, n_iter=20, cv=3, verbose=2,
                               random_state=42, n_jobs=-1)
gb_random.fit(X_train, Y_train)
print("Best parameters found by RandomizedSearchCV for Gradient Boosting:")
print(gb_random.best_params_)
Y_pred_gb_tuned = gb_random.best_estimator_.predict(X_test)
mse_gb_tuned = mean_squared_error(Y_test, Y_pred_gb_tuned)
print("Mean Squared Error (Tuned) for Gradient Boosting:", mse_gb_tuned)
print('Predicted:', round(gb_random.predict(X.iloc[0:1])[0], 1), '\n',
      'Real one:', df.loc[0, 'IMDB_Rating'])
```
